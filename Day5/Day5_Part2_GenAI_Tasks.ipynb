## Task 1: Text Generation

Generate creative content like stories, poems, or essays from a prompt.

```python
response = llm.invoke([
    SystemMessage(content="You are a creative writer."),
    HumanMessage(content="Write a short 4-line poem about artificial intelligence.")
])

print(response.content)
```

---
## Task 2: Summarization

Condense a long piece of text into a shorter version while retaining key information.

```python
long_text = """
Artificial Intelligence (AI) is a branch of computer science that aims to create 
intelligent machines that can perform tasks that typically require human intelligence. 
These tasks include visual perception, speech recognition, decision-making, and 
language translation. AI can be categorized into narrow AI, which is designed for 
specific tasks, and general AI, which would have the ability to understand and learn 
any intellectual task that a human can. Machine learning is a subset of AI that uses 
statistical techniques to give computers the ability to learn from data without being 
explicitly programmed. Deep learning, a subset of machine learning, uses neural 
networks with many layers to analyze complex patterns in data.
"""

response = llm.invoke([
    SystemMessage(content="You are a helpful assistant. Summarize the following text in 2-3 sentences."),
    HumanMessage(content=long_text)
])

print("Summary:\n", response.content)
```

## Task 3: Translation

Translate text from one language to another.

```python
response = llm.invoke([
    SystemMessage(content="You are a translator. Translate the following English text to French."),
    HumanMessage(content="Machine learning is transforming the world.")
])

print("French:", response.content)
```

## Task 4: Question Answering

Answer questions based on provided context or general knowledge.

```python
context = """
The Transformer architecture was introduced in the paper 'Attention Is All You Need' 
by Vaswani et al. in 2017. It replaced recurrent layers with self-attention mechanisms, 
enabling much faster training. Transformers are the foundation of models like BERT, GPT, 
and Mistral.
"""

response = llm.invoke([
    SystemMessage(content=f"Answer the question based on this context:\n{context}"),
    HumanMessage(content="What did the Transformer architecture replace?")
])

print(response.content)
```

## Task 5: Code Generation

Generate code from a natural language description.

## Task 6: Sentiment Analysis

Detect the emotional tone (positive, negative, neutral) in text.

```python
reviews = [
    "This product is amazing! Best purchase I've ever made.",
    "Terrible quality. Broke after one day.",
    "It's okay, nothing special but does the job."
]

for review in reviews:
    response = llm.invoke([
        SystemMessage(content="Classify the sentiment of the following text as Positive, Negative, or Neutral. Reply with only the label."),
        HumanMessage(content=review)
    ])
    print(f"Review: {review[:50]}...")
    print(f"Sentiment: {response.content}\n")
```

## Task 7: Text Classification

Categorize text into predefined labels (e.g., topic classification, spam detection).

```python
headlines = [
    "Stock market hits record high amid tech rally",
    "New study reveals benefits of daily exercise",
    "India wins cricket World Cup after thrilling final",
    "SpaceX launches new satellite into orbit"
]

for headline in headlines:
    response = llm.invoke([
        SystemMessage(content="Classify the following news headline into one category: Sports, Technology, Health, Finance, Science. Reply with only the category."),
        HumanMessage(content=headline)
    ])
    print(f"Headline: {headline}")
    print(f"Category: {response.content}\n")
```

## Task 8: Chatbot / Multi-turn Conversation

Maintain context across multiple messages in a conversation.

```python
# Multi-turn conversation — the model remembers context
messages = [
    SystemMessage(content="You are a helpful AI tutor for computer science students."),
    HumanMessage(content="What is a neural network?"),
]

response1 = llm.invoke(messages)
print("AI:", response1.content)
print("\n" + "="*60 + "\n")

# Continue the conversation
messages.append(AIMessage(content=response1.content))
messages.append(HumanMessage(content="Can you give me a simple real-world example?"))

response2 = llm.invoke(messages)
print("AI:", response2.content)
```

### Exercise 8

Create a **3-turn conversation** where you ask the model to explain **what is Python**, then ask a **follow-up question**, then ask it to **give an example**.

```python
# Exercise 8: Multi-turn Chatbot
# TODO: Build a 3-turn conversation

messages = [
    SystemMessage(content="You are a helpful tutor."),
    HumanMessage(content="YOUR FIRST QUESTION"),
]

# Turn 1
r1 = llm.invoke(messages)
print("Turn 1:", r1.content)

# Turn 2 - add context and follow-up
messages.append(AIMessage(content=r1.content))
messages.append(HumanMessage(content="YOUR FOLLOW-UP QUESTION"))
r2 = llm.invoke(messages)
print("\nTurn 2:", r2.content)

# Turn 3
messages.append(AIMessage(content=r2.content))
messages.append(HumanMessage(content="YOUR THIRD QUESTION"))
r3 = llm.invoke(messages)
print("\nTurn 3:", r3.content)
```

## Task 9: Data Extraction (Named Entity Recognition)

Extract structured information (names, dates, locations, etc.) from unstructured text.

```python
text = """
Sundar Pichai, CEO of Google, announced at the I/O conference in Mountain View 
on May 14, 2024 that Gemini AI would be integrated into all Google products.
"""

response = llm.invoke([
    SystemMessage(content="Extract all named entities from the text. Return them as: Person, Organization, Location, Date, Product. Format as a list."),
    HumanMessage(content=text)
])

print(response.content)
```

## Task 10: Paraphrasing / Rewriting

Rewrite text in different words while preserving the original meaning.

```python
original = "Machine learning algorithms can identify patterns in large datasets that humans might miss."

response = llm.invoke([
    SystemMessage(content="Paraphrase the following sentence in simpler language suitable for a beginner."),
    HumanMessage(content=original)
])

print("Original   :", original)
print("Paraphrased:", response.content)
```

## Summary

| Task | What it does | Key technique |
|------|-------------|---------------|
| Text Generation | Creates new content | Creative prompting |
| Summarization | Shortens text | Instruction-based compression |
| Translation | Language conversion | System message as translator |
| Question Answering | Answers from context | Context injection |
| Code Generation | Writes code | Natural language → code |
| Sentiment Analysis | Detects emotion | Classification prompting |
| Text Classification | Categorizes text | Label-constrained output |
| Chatbot | Multi-turn dialogue | Message history |
| Data Extraction | Pulls structured data | NER prompting |
| Paraphrasing | Rewrites text | Style-controlled generation |

**All tasks use the same model** — the difference is in **how we prompt** it!
